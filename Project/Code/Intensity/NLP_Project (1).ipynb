```python
#preproce
test_data=['The product is amazing and surprising','The kindle is great and makes my reading comfortable','The product is not worth the price']
copy_text=test_data
Text=[]
temp=[]
for i in range(len(copy_text)):
  sentence=copy_text[i].split()
  temp=[]
  for j in range(len(sentence)):
    word=sentence[j].lower()
    if word not in string.punctuation and word.isalpha()==True and word in words.words() and word not in stopwords.words():
      for token in nlp(word):
        if token.tag_!='VBZ' and token.tag_!='DT' and token.tag_!='LS' and token.tag_!='VBD' and token.tag_!='VBP' and token.tag_!='PRP' and token.tag_!='PRP$' and token.tag_!='NNP' and token.tag_!='NNS' and token.tag_!='NN'  :
          temp.append(word)
  temp=list(set(temp))
  Text.append(temp)
test_data=Text
test_data
```

```python
count1=0
count2=0
count3=0
count4=0
count5=0
intensity=[]
for sen in test_data:
  for word in sen:
    
    for check_words in sup_neg_sen:
      if word==check_words:
        count1+=1

    for check_words in neg_sen:
      if word==check_words:
        count2+=1

    for check_words in neu_sen:
      if word==check_words:
        count3+=1

    for check_words in pos_sen:
      if word==check_words:
        count4+=1

    for check_words in sup_pos_sen:
      if word==check_words:
        count5+=1
    
  inten=((5*count5+4*count4)-(3*count3+2*count2+1*count1))/(5*(5*count5+4*count4+3*count3+2*count2+1*count1))
  intensity.append(inten)
  count1=0
  count2=0
  count3=0
  count4=0
  count5=0
intensity
```
